```python
%matplotlib notebook
import numpy as np
from numba import jit
from matplotlib import pyplot as plt
from matplotlib import colors
```

```python
from ipywidgets import interact # see http://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html
import ipywidgets as w
```

```python
x_max = 1.5
y_max = 1.5
m = 5
n = 7
dim = 512

x = np.linspace(x_max * -1, x_max, dim)
y = np.linspace(y_max * -1, y_max, dim)
xs, ys = np.meshgrid(x,y,sparse=True)

ar_chladni = (np.sin(m * xs * np.pi) * np.sin(m * ys * np.pi) - np.sin(n * xs * np.pi) * np.sin(n * ys * np.pi))


        
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax.set_aspect('equal')
ax2.set_aspect('equal')
levels = (-2,-1,0,1,2)
c = ax.contour(x, y, ar_chladni,cmap='jet',levels=levels)
h = ax2.contourf(x,y, ar_chladni,cmap='jet',levels=levels)
#fig.colorbar(h,ax=ax)
#fig.show
```

```python
def update(new_m,new_n):
    ar_chladni = (np.sin(new_m * xs * np.pi) * np.sin(new_m * ys * np.pi) - np.sin(new_n * xs * np.pi) * np.sin(new_n * ys * np.pi))
    ax.clear()
    ax2.clear()
    c = ax.contour(x, y, ar_chladni,cmap='jet', levels=levels)
    h = ax2.contourf(x,y, ar_chladni,cmap='jet',levels=levels)
    ax.draw
    ax2.draw
    
m_slider = w.IntSlider(min=-10,max=24,step=1,value=m)
n_slider = w.IntSlider(min=-10,max=24,step=1,value=n)
interact(update, new_m = m_slider, new_n = n_slider)
```
